```python
import pandas as pd
import numpy as np
import copy, time
import warnings
warnings.filterwarnings('ignore')
from sklearn import datasets, metrics
from IPython.display import display
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import cross_val_score, GridSearchCV, train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from collections import defaultdict
from sklearn import datasets, metrics
from catboost import Pool, CatBoostRegressor

login = pd.read_csv('login.csv')
purchase_detail = pd.read_csv('purchase_detail.csv')
submission = pd.read_csv('submission.csv')
user_info = pd.read_csv('user_info.csv')
user_label_train = pd.read_csv('user_label_train.csv')
```

```python
#處理login, merge user_info, login
login = login.groupby("userid").agg("sum")

data = user_info
data = pd.merge(data,login, on='userid')
```

```python
#處理purchase_detail，合併
purchase_detail_g = purchase_detail.groupby("userid").agg("sum")
data = data.merge(purchase_detail_g, on="userid", how='left')
data

```

```python
#資料處理
#grass_date_count:購買次數
data["grass_date_count"] = purchase_detail["grass_date"].groupby(purchase_detail["userid"]).count()
data = data.drop("enroll_time", axis=1)
data
```

```python
data.isnull().sum()
```

```python
#有label的挑出來成為training data
train = user_label_train.merge(data, on="userid", how='left')
train
```

```python
#train填補缺失
train["birth_year"] = train["birth_year"].fillna(train["birth_year"].median())
train["gender"] = train["gender"].fillna(0)
train["grass_date_count"] = train["grass_date_count"].fillna(0)
```

```python
#增加aldult feature
f = lambda x : 1 if 2020 - float(x) >=20 else 0
aldult = train["birth_year"].apply(f)
train["aldult"] = aldult
```

```python
#建測試檔
test = submission.merge(data, on="userid", how='left')

#test填補缺失
test["birth_year"] = test["birth_year"].fillna(train["birth_year"].median())
test["gender"] = test["gender"].fillna(0)
test["grass_date_count"] = test["grass_date_count"].fillna(0)

#增加aldult feature
f = lambda x : 1 if 2020 - float(x) >=20 else 0
aldult = test["birth_year"].apply(f)
test["aldult"] = aldult


```

```python
X = train.drop(["label", "userid"], axis=1)
Y = train["label"]
testX =  test.drop(["userid"], axis=1)

trainX, valX, trainY, valY = train_test_split(X, Y, test_size=0.05, random_state=42)

scaler = StandardScaler()
#scaler = MinMaxScaler(feature_range = (0, 1))
scaler = scaler.fit(trainX)

trainX = pd.DataFrame(data=scaler.transform(trainX), columns=trainX.columns, index=trainX.index)
valX = pd.DataFrame(data=scaler.transform(valX), columns=valX.columns, index=valX.index)


testX = pd.DataFrame(data=scaler.transform(testX), columns=testX.columns, index=testX.index)

```

```python
#model = CatBoostRegressor()
model = RandomForestClassifier(n_estimators=100, min_samples_split=2, min_samples_leaf=1, 
                            max_features='sqrt', max_depth=6, bootstrap=True)

model.fit(trainX, trainY)

#pred_val = model.predict_proba(valX)[:,1]
print(cross_val_score(model, valX, valY, cv=10).mean())

```

```python
pred = model.predict_proba(testX)[:,1]

submission["label"] = pred

submission.to_csv('output.csv')

```
